```python
import pandas as pd

PATH = "/kaggle/input/competitions/pokemon-tcg-ai-battle-challenge-strategy"

# ==========================================================
# LOAD DATA
# ==========================================================

en_cards = pd.read_csv(f"{PATH}/EN_Card_Data.csv")
jp_cards = pd.read_csv(f"{PATH}/JP_Card_Data.csv")

print(f"EN cards loaded: {len(en_cards):,}")
print(f"JP cards loaded: {len(jp_cards):,}")

# ==========================================================
# COLUMN NAMES
# ==========================================================

ID_COL = "Card ID"
NAME_COL = "Card Name"

print("\nColumns:")
print(en_cards.columns.tolist())

# ==========================================================
# BUILD LOOKUP TABLES
# ==========================================================

name_to_id = {}
id_to_name = {}

for _, row in en_cards.iterrows():
    try:
        card_id = int(row[ID_COL])
        card_name = str(row[NAME_COL]).strip()

        name_to_id[card_name.lower()] = card_id
        id_to_name[card_id] = card_name

    except:
        pass

print(f"\nLoaded {len(name_to_id):,} card mappings")

# ==========================================================
# CARD SEARCH HELPER
# ==========================================================

def find_card_id(card_name):
    card_name = card_name.strip().lower()

    if card_name in name_to_id:
        return name_to_id[card_name]

    matches = [
        (cid, name)
        for cid, name in id_to_name.items()
        if card_name in name.lower()
    ]

    if matches:
        print(f"\nPossible matches for '{card_name}':")
        for cid, name in matches[:20]:
            print(f"{cid:<6} {name}")
        return matches[0][0]

    print(f"\nNo match found for '{card_name}'")
    return None

# ==========================================================
# DECK CARD IDS
# ==========================================================

TARGET_CARDS = [
    "Charmander",
    "Charmeleon",
    "Mega Charizard X ex",
    "Ultra Ball",
    "Rare Candy",
    "Master Ball",
    "Night Stretcher",
    "Energy Retrieval",
    "Switch",
    "Carmine",
    "Larry's Skill",
    "Boss's Orders",
    "Judge",
    "Brock's Scouting",
    "Cyrano",
]

CARD_IDS = {}

print("\n" + "=" * 80)
print("CARD LOOKUP")
print("=" * 80)

for card in TARGET_CARDS:
    cid = find_card_id(card)

    if cid is not None:
        CARD_IDS[card] = cid

print("\n" + "=" * 80)
print("FOUND CARD IDS")
print("=" * 80)

for k, v in CARD_IDS.items():
    print(f"{k:<25} -> {v}")

# ==========================================================
# SHOW ALL CHARIZARD RELATED CARDS
# ==========================================================

print("\n" + "=" * 80)
print("CHARIZARD CARDS")
print("=" * 80)

display(
    en_cards[
        en_cards["Card Name"]
        .str.contains("Charizard", case=False, na=False)
    ][["Card ID", "Card Name", "HP", "Category"]]
)

# ==========================================================
# SHOW ALL SUPPORTERS USED IN DECK
# ==========================================================

print("\n" + "=" * 80)
print("SUPPORTER SEARCH")
print("=" * 80)

supporters = [
    "Carmine",
    "Larry",
    "Boss",
    "Judge",
    "Brock",
    "Cyrano"
]

for s in supporters:
    print(f"\n{s}")
    display(
        en_cards[
            en_cards["Card Name"]
            .str.contains(s, case=False, na=False)
        ][["Card ID", "Card Name", "Category"]]
    )

# ==========================================================
# SAVE FOR ENV.PY
# ==========================================================

print("\n\nGenerated CARD_IDS dictionary:\n")
print("CARD_IDS = {")

for k, v in CARD_IDS.items():
    print(f'    "{k.upper().replace(" ", "_").replace("\'", "")}": {v},')

print("}")
```

```python
from dataclasses import dataclass
import pandas as pd
import numpy as np

PATH = "/kaggle/input/competitions/pokemon-tcg-ai-battle-challenge-strategy"

df = pd.read_csv(f"{PATH}/EN_Card_Data.csv")

CARD_IDS = {
    "FIRE_ENERGY": 2,
    "CHARMANDER": 926,
    "CHARMELEON": 927,
    "MEGA_CHARIZARD_X_EX": 790,
    "ULTRA_BALL": 1121,
    "RARE_CANDY": 1079,
    "ENERGY_RETRIEVAL": 1118,
    "NIGHT_STRETCHER": 1097,
    "SWITCH": 1123,
    "MASTER_BALL": 1125,
    "CARMINE": 1192,
    "LARRY_SKILL": 1206,
    "BOSS_ORDERS": 1182,
    "JUDGE": 1213,
    "BROCK_SCOUTING": 1210,
    "CYRANO": 1205,
}

@dataclass
class CardMeta:
    card_id: int
    name: str
    category: str
    stage: str
    hp: int
    evolves_from: int
    attack_cost: int
    attack_damage: int
    attack_name: str
    prizes: int

CARD_DB = {}

name_to_id = {
    str(row["Card Name"]).strip(): int(row["Card ID"])
    for _, row in df.iterrows()
}

for _, row in df.iterrows():

    card_id = int(row["Card ID"])
    name = str(row["Card Name"])

    stage_raw = str(row["Stage (Pokémon)/Type (Energy and Trainer)"])

    if "Basic Pokémon" in stage_raw:
        stage = "basic"
    elif "Stage 1 Pokémon" in stage_raw:
        stage = "stage1"
    elif "Stage 2 Pokémon" in stage_raw:
        stage = "stage2"
    else:
        stage = None

    hp = 0 if pd.isna(row["HP"]) else int(row["HP"])

    evolves_from = None

    if not pd.isna(row["Previous stage"]):
        prev_name = str(row["Previous stage"]).strip()
        evolves_from = name_to_id.get(prev_name)

    damage = 0

    if not pd.isna(row["Damage"]):
        dmg = str(row["Damage"])

        digits = "".join(c for c in dmg if c.isdigit())

        if digits:
            damage = int(digits)

    cost = 0

    if not pd.isna(row["Cost"]):
        cost = str(row["Cost"]).count("{")

    prizes = 2 if " ex" in name.lower() else 1

    CARD_DB[card_id] = CardMeta(
        card_id=card_id,
        name=name,
        category=str(row["Category"]),
        stage=stage,
        hp=hp,
        evolves_from=evolves_from,
        attack_cost=cost,
        attack_damage=damage,
        attack_name="" if pd.isna(row["Move Name"]) else str(row["Move Name"]),
        prizes=prizes,
    )

print("Cards loaded:", len(CARD_DB))

for cid in [926, 927, 790]:
    print()
    print(CARD_DB[cid])
```

```python
from dataclasses import dataclass
import random

DECK_LIST = {
    CARD_IDS["CHARMANDER"]: 4,
    CARD_IDS["CHARMELEON"]: 2,
    CARD_IDS["MEGA_CHARIZARD_X_EX"]: 3,
    CARD_IDS["ULTRA_BALL"]: 4,
    CARD_IDS["RARE_CANDY"]: 4,
    CARD_IDS["MASTER_BALL"]: 2,
    CARD_IDS["NIGHT_STRETCHER"]: 2,
    CARD_IDS["ENERGY_RETRIEVAL"]: 2,
    CARD_IDS["SWITCH"]: 2,
    CARD_IDS["CARMINE"]: 4,
    CARD_IDS["LARRY_SKILL"]: 3,
    CARD_IDS["BOSS_ORDERS"]: 3,
    CARD_IDS["JUDGE"]: 2,
    CARD_IDS["BROCK_SCOUTING"]: 3,
    CARD_IDS["CYRANO"]: 2,
    CARD_IDS["FIRE_ENERGY"]: 18,
}

assert sum(DECK_LIST.values()) == 60


@dataclass
class PokemonInPlay:
    card_id: int
    current_hp: int
    attached_energy: int = 0


@dataclass
class PlayerState:
    deck: list
    hand: list
    discard: list
    prizes: list
    active: object = None
    bench: list = None
    prizes_taken: int = 0
    supporter_played: bool = False

    def draw(self, n=1):
        for _ in range(n):
            if self.deck:
                self.hand.append(self.deck.pop(0))


def make_deck():
    deck = []

    for cid, cnt in DECK_LIST.items():
        deck.extend([cid] * cnt)

    random.shuffle(deck)

    return deck


def setup_player():

    deck = make_deck()

    hand = []

    for _ in range(7):
        hand.append(deck.pop())

    prizes = []

    for _ in range(6):
        prizes.append(deck.pop())

    return PlayerState(
        deck=deck,
        hand=hand,
        discard=[],
        prizes=prizes,
        bench=[None] * 5
    )


player = setup_player()

print("Deck:", len(player.deck))
print("Hand:", len(player.hand))
print("Prizes:", len(player.prizes))
```
